```matlab
imdata = imread('https://images0.cnblogs.com/blog/614265/201408/241708117531878.png');
I=im2bw(imdata);
I=I+0;
imshow(I)
```

```matlab
r = normrnd(0,1,size(I));
J=I+0;
J(find(J==0))=-1;
J = J+r;
imshow(J)
```

```matlab
[M,N]=size(J);
mu = J;
for t= 1 : 10
    for i = 1:M
        for j = 1:N
            m(i,j) = sum(size(linju(i,j,M,N,J))*mu(i,j));
            mu(i,j) =tanh(m(i,j)+1/2*(L(sum(linju(i,j,M,N,mu))>0)-L(sum(linju(i,j,M,N,mu))<0)));
        end
    end
end
figure
imshow(mu)
```
