```matlab
imdata = imread('https://images0.cnblogs.com/blog/614265/201408/241708117531878.png');
I=im2bw(imdata);
I=I+0;
imshow(I)
```

# Guassian noise

```matlab
r = normrnd(0,1,size(I));
J=I+0;
J(find(J==0))=-1;
J = J+r;
imshow(J)
```

```matlab
[M,N]=size(J);
T=J;
J=T;
for i = randperm(M,M)
    for j = randperm(N,N)
        if E(i,j,1,linju(i,j,M,N,J),J)<E(i,j,-1,linju(i,j,M,N,J),J)
            J(i,j)=1;
        else
            J(i,j)=-1;
        end
    end
end
figure
imshow(J)
```

# pepper noise

```matlab
J = imnoise(I,'salt & pepper',0.2);
imshow(J)
```

```matlab
[M,N]=size(J);
T=J;
J=T;
for i = randperm(M,M)
    for j = randperm(N,N)
        if E(i,j,1,linju(i,j,M,N,J),J)<E(i,j,-1,linju(i,j,M,N,J),J)
            J(i,j)=1;
        else
            J(i,j)=-1;
        end
    end
end
figure
imshow(J)
```
